````python
import numpy as np
import tensorflow as tf #face bk propagationa+ layere
print("TransformFlow version:", tf.__version__)
````

````python
x=np.array([[1, 1], [1,0], [0,1], [0,0]])
y=np.array([0.0, 1.0, 1.0, 0.0])
model =tf.keras.models.Sequential([
    tf.keras.layers.Dense(2, input_dim=2, activation='sigmoid'),
                                tf.keras.layers.Dense(1, activation='sigmoid'),   
                                   
                                   ])
````

````python
x
````

````python
y
````

````python
model
````

````python
#compilam cu un optimizator!!!
#putm folosi gradient descen - sgd ; adam e mai puternic

#ca functie de loss = avem BinaryCrossentropy = negativelikly hood = 0 sau 1

#metrica = p a vedea dupa fiecare antrebnare: acuratete, precizie..
model.compile(optimizer ="adam",
              loss=tf.keras.losses.BinaryCrossentropy(),
              metrics=['accuracy']
              )
````

# Predictia mdelului- cu Wait random(neantrenat)

````python
#vedem un model care nu a fost antrenat, avem weit-uri random
predictions = model(x)
predictions
````

# Antrenarea modelului
**model.fit (date intrare, date de iesire dorite, nr de antrenament* 


epochs = dupa fiecare epoca, ii cresre acuratetea, deoarece:GRADIENT DESCEND, merge catre minimul global
dupa fiecare epoca, loss scade==>e mai aproape de a gasi regula din spate 

````python
model.fit(x, y, epochs=1000)
````

#Dupa antrenarea in 1000 de epoci, evaluama *X*

````python
model.evaluate(x, y, verbose=2)
````
